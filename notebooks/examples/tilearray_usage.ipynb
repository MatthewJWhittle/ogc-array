# OGC Array Library Usage Example

This notebook demonstrates how to use the OGC Array library to fetch geospatial data from OGC services and create visualizations.

## Features Demonstrated:
- Fetching data from WCS services
- Creating Dask-backed xarray objects
- Plotting geospatial data
- Working with different coordinate systems


```python
import tilearray as ta
import xarray as xr
from tilearray.types import CRS
from tilearray.service.config import WCSConfig
from tilearray.types import Format
```

```python
url = "https://environment.data.gov.uk/spatialdata/lidar-composite-digital-terrain-model-dtm-1m/wcs"


coverage_id = "13787b9a-26a4-4775-8523-806d13af58fc__Lidar_Composite_Elevation_DTM_1m"
width = 4096
xmin, ymin  = 431900.0, 382700.0
xmax = xmin + width
ymax = ymin + width
bbox = (xmin, ymin, xmax, ymax)
chunk_size = 256, 256
res = 5.0
config = WCSConfig.from_url(
    url,
    coverage_id=coverage_id,
    crs=CRS.EPSG_27700,
    output_format=Format.GEOTIFF,
    chunk_size=chunk_size,
    resolution=(res, res)
)

result = ta.create_array(
    service_url=config,
    bbox=bbox,
    crs=CRS.EPSG_27700,
)

data = result.compute()  # sequential requests for stability

data.plot()
```
